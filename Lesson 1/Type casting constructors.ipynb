# Type casting constructors"

# In Python, "type casting constructors" refer to the built-in functions that allow explicit conversion of a value from one data type to another.

# These functions act as constructors for the target data type, creating a new object of that type based on the provided input.

# 1. Implicit Type Conversion

```python
# Python automatically converts
# a to int
a = 7
print(type(a))
```

```python
# Python automatically converts
# b to float
b = 3.0
print(type(b))
```

```python
# Python automatically converts
# c to float as it is a float addition
c = a + b
print(c)
print(type(c))
```

```python
# Python automatically converts
# d to float as it is a float multiplication
d = a * b
print(d)
print(type(d))
```

# 2. Explicit Type Conversion

# int to Float datatype

```python
A = 5

# typecast to float
B = float(A)

print(B)
print(type(B))
```

# Float to int datatype

```python
a = 5.9

# typecast to int
n = int(a)

print(n)
print(type(n))
```

# int to String datatype

```python
a = 5

# typecast to str
n = str(a)

print(n)
print(type(n))
```

# string into float datatype

```python
a = "5.9"

# typecast to float
n = float(a)

print(n)
print(type(n))
```

# string to int datatype

```python
a = "5"
b = 't'

# typecast to int
n = int(a)

print(n)
print(type(n))

print(int(b))
print(type(b))
```
