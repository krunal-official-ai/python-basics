# 3. Logical Errors

# Logical error occurs when the program produces incorrect results.
# Logical error → Code runs but gives the wrong answer.

# For example,
# let's say we are creating a simple program to convert
# temperature from Fahrenheit to Celsius

```python
# Calculate average of 3 numbers
a, b, c = 10, 20, 30
average = a + b + c / 3   # ❌ logical error
print("Average:", average)
```

# But the correct average should be 20.0.
# The error is due to operator precedence – division is happening before addition.

```python
a, b, c = 10, 20, 30
average = (a + b + c) / 3
print("Average:", average)
```
