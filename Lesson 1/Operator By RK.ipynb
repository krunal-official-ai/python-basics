# Bitwise Operator
As the name goes by, bitwise operators perform operations bit by bit.

```python
# Bitwise AND operator: Returns 1 if both the bits are 1 else 0.
#Example:

a = 10 = 1010 (Binary)
b = 4 =  0100 (Binary)

a & b = 1010
         &
        0100
      = 0000
      = 0 (Decimal)
```

```python
# Bitwise OR operator: Returns 1 if Any one the bits are 1 else 0.
x = 10 = 1010
y = 4 = 0100

x | y = 1010
         |
        0100
      = 1110
      =  14 (Decimal)
```

```python
# Bitwise not operator: Returns one’s complement of the number. [ ~ : approximately]
# Example:

a = 10 = 1010 (Binary)

~a = ~1010 #(~ tilde)
   = -(1010 + 1)
   = -(1011)
   = -11 (Decimal)

x = 5
print(~x)
x = ~5
  = -5 - 1
  = -6       # Output: -6   # binary:  00000101


x = -2
print(~x)
x = ~(-2)
  = -(-2) - 1
  = 2 - 1
  = 1        # 00000001
```

```python
Bitwise xor operator: Returns 1 if one of the bits is 1 and theb bot are 1s or 0th then the answer is 0.
Example:

a = 10 = 1010 (Binary)
b = 4 =  0100 (Binary)

a ^ b = 1010 #(^ Caret)
         ^
        0100
      = 1110
      = 14 (Decimal)
```

# Membership Operators

```python

a = 5
b = 10
list = [1, 2, 3, 4, 5 ];

if ( a in list ):
   print ("a is available in the given list")

else:
   print ("a is not available in the given list")
```
