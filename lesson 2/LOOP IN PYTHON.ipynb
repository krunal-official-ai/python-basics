# In Python, loops are used to repeat a block of code multiple times. There are two main types of loops:

# 🔁 for Loop
# 🔁 while Loop

# 🔁 for loop

# # A for loop in Python is used to iterate over a sequence
# # like a list, tuple, string, dictionary, or range and run a block of code once for each item in that sequence.


# # Consider the following trivial problem:

# # Let's say we want to,
# print string "I am Ai Expert" 100 times to the console.
# One way to achieve this is to create a Python script and call print() function 100 times as follows:

```python
print("I am Ai Expert")  # 1st time
print("I am Ai Expert")  # 2nd time
...
...
print("I am Ai Expert")  # 100th time
```

# ✅ Syntax:

```python
for variable in sequence:
    # code block
```

# -- variable : a name that takes the value of each item in the sequence.
# -- sequence : Something you can loop through (like a list, string, or range)

```python
Basket = ['Apple', 'Banana', 'Cherry']
print(Basket)

for fruit in Basket:
    print(fruit)
```

# ✅ The range() function : 
# -- it used to generate a sequence of numbers, often used in for loops.

# range(start, stop, step)

# | `start` (Optional) Starting number (default is `0`)
# | `stop`  Required. Ends **before** this number                 
# | `step`  (Optional) Difference between each number (default is `1`)


```python
for i in range(100):
    print("I am Ai Expert",i)
```

# ✅ With start and stop

```python
for i in range(1, 10):
    print(i)
```

# ✅ With step

```python
for i in range(1, 100, 5):
    print(i)
```

# ✅ Countdown (using negative step)

```python
for i in range(5, 0, -1):
    print(i)
```

# ✅ CODE : Loop through a range

```python
for i in range(5):
    print(i)
```

# ✅ CODE : Loop through a list

```python
fruits = ["apple", "banana", "cherry"]
for X in fruits:
    print(X)
```

# 🔁 2. while Loop

# A while loop in python continues to execute the code as long as the given condition is true, if it becomes false, then the loop stops.

# Another reason to use while loop is when we aren’t sure about how many times to iterate.

# Repeats as long as a condition is True.

```python
i = 2
while i <= 200:
    print("Today is Funday",i)
    i += 1    # i = i+1
```

# ✅ Example:

```python
i = 1
while i <= 5:
    print(i)
    i += 1 # i= i+1
```

# ✅ while Loop with break

```python
i = 1
while True:
    print(i)
    if i == 10:
        break
    i += 1
```

# >> Stops when i becomes 10.

# ✅ while Loop with continue

```python
i = 0
while i < 10:
    i += 1
    if i == 7:
        continue
    print(i)
```

# >> Skips printing 7.

# ✅ User Input Until Correct Answer

```python
password = ""
while password != "bhavesh":
    password = input("Enter the password: ")

print("Access granted!")
```

# ✅ Countdown Timer

```python
count = int(input("Enter your Num: "))
while count > 0:
    print("Your Rocket Countdown : ", count)
    count -= 1

print("Liftoff!")
```

# ✅ Sum of First N Numbers

```python
n = 10
i = 1
total = 0
while i <= n:
    total += i
    i += 1
print("Sum is:", total) #1+2+3+4+5 = 15
```

# 🧮 Example : Simple Menu System (Real-World App)

```python
while True:
    print("\n--- Menu ---")
    print("1. Say Hello")
    print("2. Show Help")
    print("3. Exit")

    choice = input("Enter your choice (1-3): ")

    if choice == "1":
        print("Hello! 👋")
    elif choice == "2":
        print("Help: Choose 1 to greet, 3 to exit.")
    elif choice == "3":
        print("Goodbye!")
        break
    else:
        print("Invalid choice. Try again.")
```

# 🔁 Example : Nested while Loop (Multiplication Table)

```python
i = 1
while i <= 3:
    j = 1
    while j <= 3:
        print(f"{i} x {j} = {i * j}")
        j += 1
    print("---")
    i += 1
```

# 🎮 Example : Guess the Number Game

```python
import random

secret = random.randint(1, 10)
guess = 0

while guess != secret:
    guess = int(input("Guess the number (1-10): "))

    if guess < secret:
        print("Too low!")
    elif guess > secret:
        print("Too high!")
    else:
        print("Correct! 🎉")
```

# 🔐 Example : Login System with Limited Attempts

```python
username = "admin"
password = "1234"
attempts = 0

while attempts < 3:
    user = input("Username: ")
    pwd = input("Password: ")

    if user == username and pwd == password:
        print("Login successful ✅")
        break
    else:
        print("Incorrect credentials.")
        attempts += 1

if attempts == 3:
    print("Too many failed attempts. Access blocked.")
```

# 🔁 Loop Control Statements

# | `break`  | Exit the loop |
# | `continue` | Skip the current iteration |

# 🔁 break :

# The break statement is used to exit a loop early,
# even if the loop condition is still True. It is commonly used when a certain condition is met inside the loop body.



```python
while condition:
    # code block
    if some_condition:
        break  # exit the loop
    # more code
```

```python
for i in range(10):
    if i == 4:
        break
    print(i)
```

```python
i = 1
while i <= 10:
    print(i)
    if i == 7:
        break  # loop will stop when i becomes 5
    i += 1
```

```python
while True:
    user_input = input("Enter something (or type 'exit' to quit): ")
    if user_input == "exit":
        print("Loop exited.")
        break
    print("You typed:", user_input)
```

# 🔁 continue :

# The continue statement is used to skip the rest of the code inside the loop for the current iteration and go back to the top of the loop for the next iteration.



```python
while condition:
    # some code
    if some_condition:
        continue  # skip the rest and go to next iteration
    # more code (skipped if 'continue' is hit)
```

```python
for i in range(5):
    if i == 2:
        continue
    print(i)
```

```python
i = 0
while i < 10:
    i += 1
    if i == 8:
        continue  # skip printing 5 number
    print(i)
```

```python
i = 0
while i < 10:
    i += 1
    if i % 2 == 0:
        continue
    print(i)
```
