# In Python,
# Conditional statements are used to perform different actions based on different conditions.
# The most common conditional statements are:

# 🎯 if Statement

```python
if condition/expression:
    # code block to execute if condition is True
```


# If the expression returns True, then the block code (conditional statement) runs, & if the statement is false, the expression is returned False and the program ends without running the conditional statement.

```python
X = 7
if X > 5:
    print("x is greater than 5")
```

# 🎯 if...else Statement
# Syntax :

```python
if expression:
    # ..........statement
else:
    # ..........statement
```


# This statement is use to check whether the condition is true or not however { if } the condition doesn’t meet the criteria then the { else } statement gets executed.

```python
x = 7
if x >= 2:
    print("greater")
else:
    print("less")
```

```python
balance = 1500
withdraw = 2000

if withdraw <= balance:
    print("Withdrawal successful 💰")
else:
    print("Insufficient balance ❗")
```

```python
cart_value = 1200

if cart_value > 1000:
    discount = 0.10  # 10%
else:
    discount = 0.05  # 5%

final_amount = cart_value - (cart_value * discount)
print("Final amount to pay:", final_amount)
```

# 🎯 if...elif...else Statement

```python
# Syntax :

if expression:
    # ..........statement
elif:
    # ..........statement
elif:
    # ..........statement
else:
    # ..........statement
```

```python
x = int(input("N:"))

if x > 5:
    print("x is greater than 5")

elif x == 5:
    print("x is exactly 5")

else:
    print("x is less than 5")
```

```python
grade = int(input("Enter your Grade:"))

if  grade >= 90:
    print("Grade: A+")

elif grade >= 80:     #case1 grade >= 80:
    print("Grade: A")

elif grade >= 70:
    print("Grade: B")

elif grade >= 60:
    print("Grade: C")

elif grade >= 50:
    print("Grade: D")

elif grade >= 40:
    print("Grade: E")

else:
    print("Game over")
```

# **MATCH CASE -**

# match case in Python is just like saying:
# 👉 “If this happens — do this.”
# 👉 “If that happens — do that.”
# 👉 “Otherwise — do something else.”

# It works like a smart version of if-elif-else

```python
command = "start"

match command:
    case "start":
        print("System starting...")
    case "stop":
        print("System stopping...")
    case "restart":
        print("System restarting...")
    case _:
        print("Unknown command")
```

```python
status_code = 404

match status_code:
    case 200 | 201:
        print("Success")
    case 400 | 404:
        print("Client error")
    case 500:
        print("Server error")
    case _:
        print("Unknown status")
```

```python
number = 15

match number:
    case n if n > 0:
        print("Positive number")
    case n if n < 0:
        print("Negative number")
    case _:
        print("Zero")
```

```python
marks = 85

match marks:
    case m if m >= 90:
        print("Grade A+ – Excellent 🎯")
    case m if m >= 80:
        print("Grade A – Very Good 👏")
    case m if m >= 70:
        print("Grade B – Good 🙂")
    case m if m >= 60:
        print("Grade C – Satisfactory")
    case _:
        print("Fail ❌")
```

```python
choice = int(input("1.Deposit  2.Withdraw  3.Check Balance  4.Exit → "))

match choice:
    case 1:
        print("💰 Depositing money...")
    case 2:
        print("💸 Withdrawing money...")
    case 3:
        print("🏦 Checking account balance...")
    case 4:
        print("👋 Thank you for banking with us!")
    case _:
        print("⚠️ Invalid option, please try again.")
```

```python
status = "shipped"

match status:
    case "pending":
        print("🕓 Order pending confirmation.")
    case "packed":
        print("📦 Order is packed and ready to ship.")
    case "shipped":
        print("🚚 Order is on the way!")
    case "delivered":
        print("📬 Order delivered successfully.")
    case _:
        print("❌ Unknown status.")
```
