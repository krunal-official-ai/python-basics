# 🔢  LIST COMPREHENSION

# List comprehension in Python provides a concise and elegant way to create lists.

# It's a more readable and often more efficient alternative to using traditional for loops with append() to build lists.

# Syntax: if

```python
[expression for item in iterable if condition]
```

# Condition: This is a boolean expression.
# Only if this condition evaluates to True will the item be processed by the expression and included in the new list.

# An expression is a combination of values, variables, operators, and function calls that Python can evaluate to produce a result.

```python
even_numbers = [ ]

for num in range(10):

    if num % 2 == 0:

        even_numbers.append(num)

print(even_numbers)
```

# Using list comprehension:

```python
            #[expression for item in iterable if condition]
even_numbers = [num for num in range(10) if num % 2 == 0]

print(even_numbers)
```

# Syntax: if else

```python
[expression_if_true if condition else expression_if_false for item in iterable]
```

```python
even_odd_labels = ["Even" if num % 2 == 0 else "Odd" for num in range(5)]

print(even_odd_labels)
```

# Syntax : Nested List Comprehensions

```python
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

flat_list = [num for row in matrix for num in row]

print(flat_list)
```
