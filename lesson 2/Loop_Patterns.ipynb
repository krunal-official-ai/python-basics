# Right-Angled Triangle

```python
n = 5
for i in range(1, n + 1):
    print('* ' * i)
```

# Inverted Triangle

```python
n = 5
for i in range(n, 0, -1):
    print('* ' * i)
```

# Pyramid Pattern

```python
n = 5
for i in range(1, n + 1):
    print(' ' * (n - i) + '*' * (2 * i - 1))
```

# Diamond Pattern

```python
n = 5

# upper half
for i in range(1, n + 1):
    print(' ' * (n - i) + '*' * (2 * i - 1))


# lower half
for i in range(n - 1, 0, -1):
    print(' ' * (n - i) + '*' * (2 * i - 1))
```

# Number Triangle

```python
n = 5
for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, end=' ')
    print()
```

# Floyd's Triangle

```python
n = 5
num = 1
for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(num, end=' ')
        num += 1
    print()
```
